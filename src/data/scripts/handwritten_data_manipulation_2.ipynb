```python
import pandas as pd
```

```python
df = pd.read_csv("../clean/QA_SR_2022_Expert.csv", delimiter=";")
df
```

```python
for i in range(len(df)):
    print(f"Question: {df.loc[i, 'question']}\nAnswer: {df.loc[i, 'answer']}\nContext: {df.loc[i, 'context']}")
    print(df.loc[i, "context"].lower().find(str(df.loc[i, "answer"]).lower()), df.loc[i, "IS_SUBSTRING"])
    print()
```

```python
answers = df.apply(lambda row: {"text": [row["answer"]], "answer_start": [row["IS_SUBSTRING"]]}, axis=1)
# print (answers)
for answer in answers:
    print(answer)
```

```python
df_squad = pd.DataFrame({"question": df["question"], "context": df["context"], "answers": answers})
df_squad.to_csv("../clean/QA_SR_2022_Expert-squad-format.csv", index=False, sep=";")
df_squad
```
