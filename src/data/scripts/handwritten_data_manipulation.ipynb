```python
import pandas as pd
```

### Combine all tabs

```python
appendices_tab = pd.read_csv('../clean/sr-2022-questions-answers-appendices.csv', delimiter=';', encoding='utf-8')
print(f"{appendices_tab.shape=}")
contribution_tab = pd.read_csv('../clean/sr-2022-questions-answers-contribution.csv', delimiter=';', encoding='utf-8')
print(f"{contribution_tab.shape=}")
governance_tab = pd.read_csv('../clean/sr-2022-questions-answers-governance.csv', delimiter=';', encoding='utf-8')
print(f"{governance_tab.shape=}")
metrics_tab = pd.read_csv('../clean/sr-2022-questions-answers-metrics.csv', delimiter=';', encoding='utf-8')
print(f"{metrics_tab.shape=}")
strategy_tab = pd.read_csv('../clean/sr-2022-questions-answers-sustainability-strategy.csv', delimiter=';', encoding='utf-8')
print(f"{strategy_tab.shape=}")
finance_tab = pd.read_csv('../clean/sr-2022-questions-answers-sustainable-finance.csv', delimiter=';', encoding='utf-8')
print(f"{finance_tab.shape=}")
operations_tab = pd.read_csv('../clean/sr-2022-questions-answers-sustainable-operations.csv', delimiter=';', encoding='utf-8')
print(f"{operations_tab.shape=}")
```

```python
print(appendices_tab.loc[0, 'question'])
print(appendices_tab.loc[0, 'answer'])
print(appendices_tab.loc[0, 'context'])
appendices_tab
```

```python
# combine all the dataframes
all_tabs = pd.concat([appendices_tab, contribution_tab, governance_tab, metrics_tab, strategy_tab, finance_tab, operations_tab])
print(f"{all_tabs.shape=}")
# save to single file
all_tabs.to_csv('../clean/sr-2022-questions-answers-ALL.csv', sep=';', encoding='utf-8', index=False)
```

### Squad format

```python
start = all_tabs.apply(lambda row: row["context"].lower().find(row["answer"].lower()), axis=1)

all_tabs_squad = pd.DataFrame({"context": all_tabs["context"], "question": all_tabs["question"], "answer": all_tabs["answer"], "start": start})
```

```python
all_tabs_squad_no_answer = all_tabs_squad[all_tabs_squad["start"] == -1]
display(all_tabs_squad_no_answer)
print("Number of answers with no substring: ", len(all_tabs_squad_no_answer))
```

```python
# some answers are not substrings in context, some have issues due to formatting, some are paraphrased
for index, row in all_tabs_squad_no_answer.iterrows():
    print(f"{row['question']=}")
    print(f"{row['answer']=}")
    print(f"{row['context']=}")
    print()
```

```python
answers = [{"text": [all_tabs.iloc[i]["answer"]], "answer_start": [s]} for i, s in enumerate(start)]
answers = pd.Series(answers)
answers

all_tabs_squad_format = pd.DataFrame({"question": all_tabs["question"].reset_index(drop=True), "context": all_tabs["context"].reset_index(drop=True), "answers": answers.reset_index(drop=True)})
all_tabs_squad_format.to_csv('../clean/sr-2022-questions-answers-ALL-squad-format.csv', sep=';', encoding='utf-8', index=False)
```
